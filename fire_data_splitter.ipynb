```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import mapping, Point
import matplotlib.pyplot as plt
```

```python
def str_cut(a):
    return a[:4]
```

```python
df = pd.read_csv('./DL_FIRE_M-C61_426750/fire_archive_M-C61_426750.csv')
df2 = pd.read_csv('./DL_FIRE_M-C61_426750/fire_nrt_M-C61_426750.csv')
```

```python
# Sort df by solely California data

# Convert latitude and longitude to Points
df['geometry'] = df.apply(lambda x: Point((x.longitude, x.latitude)), axis=1)
# Convert to a GeoDataFrame
fire_gdf = gpd.GeoDataFrame(df, geometry='geometry')
# Read GeoJSON file containing California boundaries
california_boundaries = gpd.read_file('California_County_Boundaries.geojson')
# Perform spatial join
df = gpd.sjoin(fire_gdf, california_boundaries, op='within')
# Drop unnecessary columns resulting from the spatial join
df = df.drop(columns=['index_right'])
```

```python
df.head()
```

```python
# Sort df2 by solely California data

# Convert latitude and longitude to Points
df2['geometry'] = df2.apply(lambda x: Point((x.longitude, x.latitude)), axis=1)
# Convert to a GeoDataFrame
fire_gdf = gpd.GeoDataFrame(df2, geometry='geometry')
# Read GeoJSON file containing California boundaries
california_boundaries = gpd.read_file('California_County_Boundaries.geojson')
# Perform spatial join
df2 = gpd.sjoin(fire_gdf, california_boundaries, op='within')
# Drop unnecessary columns resulting from the spatial join
df2 = df2.drop(columns=['index_right','SHAPE_Length','SHAPE_Area])
```

```python

df['year'] = df['acq_date'].apply(str_cut)
df2['year'] = df2['acq_date'].apply(str_cut)
```
